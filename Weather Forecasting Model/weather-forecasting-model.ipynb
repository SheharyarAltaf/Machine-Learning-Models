# About Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/weather-dataset-rattle-package/weatherAUS.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.shape
```

```python
# as there are some columns with very little values so we are gonna delete those.
df = df.drop(["Evaporation","Sunshine","Cloud9am","Cloud3pm","Location", "Date"], axis =1)
df.head()
```

```python

df.shape
```

```python
df = df.dropna(axis = 0)
df.head()
```

# Vectorization

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['WindGustDir']=le.fit_transform(df['WindGustDir'])
```

```python
df['WindDir3pm']=le.fit_transform(df['WindDir3pm'])
df['WindDir9am']=le.fit_transform(df['WindDir9am'])
df['RainToday']=le.fit_transform(df['RainToday'])
df['RainTomorrow']=le.fit_transform(df['RainTomorrow'])
```

```python
X = df.drop(['RainTomorrow'], axis = 1)
y = df['RainTomorrow']
```

```python
X
```

# Data Visualization

```python
corrmat = df.corr(method = "spearman")
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(corrmat,annot=True)
```

```python
plt.figure(figsize = (8,8))
sns.scatterplot(x = 'MaxTemp', y = 'MinTemp', hue = 'RainTomorrow' , palette = 'inferno',data = df)
```

```python
plt.pie(df['RainTomorrow'].value_counts(), labels=['Yes','No'],autopct="%0.2f")
plt.show()
```

# Train Test Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=45)
```

# Logistic Regression

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
pred_1=model.predict(X_test)
accuracy=accuracy_score(y_test,pred_1)
print("Accuracy Score is:", accuracy)
```

# Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model_1=DecisionTreeClassifier()
model_1.fit(X_train,y_train)
pred_2=model_1.predict(X_test)
accuracy=accuracy_score(y_test,pred_2)
print("Accuracy Score is:", accuracy)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
model_2=RandomForestClassifier()
model_2.fit(X_train,y_train)
pred=model_2.predict(X_test)
accuracy=accuracy_score(y_test,pred)
print("Accuracy Score is:", accuracy)
```
